# STRANGE • Hypothesis Lab Demo

```

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../results/strange_hypothesis_lab.csv')
df.head()
```

```

# Curva de reward promedio por episodio
ep_reward = df.groupby('episode')['reward'].sum()
plt.figure()
plt.plot(ep_reward.index, ep_reward.values)
plt.xlabel('Episode')
plt.ylabel('Total reward')
plt.title('Strange • Total reward per episode')
plt.show()
```

```

# Evolución de T y gate w_t en las primeras 3 episodios
subset = df[df['episode'] < 3]
plt.figure()
plt.plot(subset['step'], subset['T'], label='T')
plt.plot(subset['step'], subset['w_t'], label='w_t')
plt.xlabel('Step')
plt.ylabel('Value')
plt.legend()
plt.title('Temperature and gate (first 3 episodes)')
plt.show()
```
