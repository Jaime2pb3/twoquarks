```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_strange = pd.read_csv('../results/strange_hypothesis_lab.csv')
df_anti    = pd.read_csv('../results/anti_strange_hypothesis_lab.csv')

print("STRANGE:")
display(df_strange.head())

print("ANTI-STRANGE:")
display(df_anti.head())
```

```python
ep_strange = df_strange.groupby('episode')['reward'].sum()
ep_anti    = df_anti.groupby('episode')['reward'].sum()

plt.figure(figsize=(10,5))
plt.plot(ep_strange.index, ep_strange.values, label='STRANGE')
plt.plot(ep_anti.index,    ep_anti.values,    label='ANTI-STRANGE')
plt.xlabel('Episode')
plt.ylabel('Total reward')
plt.title('STRANGE vs ANTI-STRANGE — Reward por episodio')
plt.legend()
plt.grid(True)
plt.show()
```

```python

div_strange = df_strange.groupby('episode')['diversity'].mean()
div_anti    = df_anti.groupby('episode')['diversity'].mean()

plt.figure(figsize=(10,5))
plt.plot(div_strange.index, div_strange.values, label='STRANGE')
plt.plot(div_anti.index,   div_anti.values,   label='ANTI-STRANGE')
plt.xlabel('Episode')
plt.ylabel('Mean diversity')
plt.title('Diversidad interna')
plt.legend()
plt.grid(True)
plt.show()
```

```python
T_strange = df_strange.groupby('episode')['T'].mean()
T_anti    = df_anti.groupby('episode')['T'].mean()

plt.figure(figsize=(10,5))
plt.plot(T_strange.index, T_strange.values, label='STRANGE T')
plt.plot(T_anti.index,   T_anti.values,   label='ANTI-STRANGE T')
plt.xlabel('Episode')
plt.ylabel('Mean Temperature')
plt.title('Temperatura media')
plt.legend()
plt.grid(True)
plt.show()
```

```python
w_strange = df_strange.groupby('episode')['w_t'].mean()
w_anti    = df_anti.groupby('episode')['w_t'].mean()

plt.figure(figsize=(10,5))
plt.plot(w_strange.index, w_strange.values, label='STRANGE gate')
plt.plot(w_anti.index,   w_anti.values,   label='ANTI-STRANGE gate')
plt.xlabel('Episode')
plt.ylabel('Gate weight')
plt.title('Gate w_t')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Mean reward:")
print("STRANGE:", np.mean(ep_strange))
print("ANTI-STRANGE:", np.mean(ep_anti))

print("\nStd reward:")
print("STRANGE:", np.std(ep_strange))
print("ANTI-STRANGE:", np.std(ep_anti))

print("\nMean diversity:")
print("STRANGE:", np.mean(div_strange))
print("ANTI-STRANGE:", np.mean(div_anti))

print("\nMean Temperature:")
print("STRANGE:", np.mean(T_strange))
print("ANTI-STRANGE:", np.mean(T_anti))
```
